```python
import pandas as pd
from rich import print
%load_ext rich
```

```python
data = pd.read_csv('../src/reasoning_eval/mt/DeepSeekUtils/outputs/run_summary_deepseek-r1_deepseek-r1.csv')
```

```python
data['golden_mqm_score'] = data['golden_mqm_score'].apply(lambda x: float(x) if x > -25 else float(-25))
```

```python
data['llm_mqm_score'] = data['llm_mqm_score'].abs()
```

```python
data['diff'] = abs(data['llm_mqm_score'].abs() - data['golden_mqm_score'].abs())
```

```python
data['reasoning_tokens'] = data['reasoning_content'].apply(lambda x: len(x.split(' ')))
```

```python
data = data.sort_values(by=['llm_mqm_score', 'reasoning_tokens'])
```

```python
def display_data(idx, df):
    print(f"""
[bold blue]Source:[/bold blue] {df.iloc[idx]['src_segment']}
[bold green]Hypothesis:[/bold green] {df.iloc[idx]['hyp_segment']}
[bold yellow]Diff:[/bold yellow] |Human([red]{df.iloc[idx]['golden_mqm_score']}[/red]) - LLM([cyan]{df.iloc[idx]['llm_mqm_score']}[/cyan])| = [magenta]{df.iloc[idx]['diff']}[/magenta]
[bold orange1]Reasoning:[/bold orange1] {df.iloc[idx]['reasoning_content']}
          """)
```

```python
display_data(-1, data)
```
